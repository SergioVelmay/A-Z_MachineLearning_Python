# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Data.csv')
# X:Y is a range from X to Y
# : is all range from first to last
# -1 is the last element
features = dataset.iloc[:, :-1].values # all rows, all columns but last
variable = dataset.iloc[:, -1].values # all rows, only last column
```

```
print(features)
```

```
print(variable)
```

## Taking care of missing data

```
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(features[:, 1:3]) # first (0) and last (3) columns are not selected
missing = imputer.transform(features[:, 1:3])
features[:, 1:3] = missing
```

```
print(missing)
```

```
print(features)
```

## Encoding categorical data

### Encoding the Independent Variable

```
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
# we want to encode France, Germany and Spain as [1,0,0], [0,1,0] and [0,0,1]
hot_encoder = OneHotEncoder()
# transform only the Country column (index=0) but return all the columns (passthrough)
coltrans = ColumnTransformer(transformers=[('encoder', hot_encoder, [0])], remainder='passthrough')
transformed = coltrans.fit_transform(features)
features = np.array(transformed)
```

```
print(transformed)
```

```
print(features)
```

### Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
variable = label_encoder.fit_transform(variable)
```

```
print(variable)
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, variable, test_size=0.2, random_state=1)
```

```
print(X_train)
```

```
print(X_test)
```

```
print(y_train)
```

```
print(y_test)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# all the rows but only numeric columns, not 1st, 2nd and 3rd columns
scaled_train = scaler.fit_transform(X_train[:, 3:])
X_train[:, 3:] = scaled_train
# transform not fit_transform to apply the same transformation scale
scaled_test = scaler.transform(X_test[:, 3:])
X_test[:, 3:] = scaled_test
```
