# Logistic Regression

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values # pandas.DataFrame.iloc (integer-location)
y = dataset.iloc[:, -1].values # based indexing for selection by position
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
# sklearn.model_selection.train_test_split
# split arrays or matrices into random train and test subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
# 400 observations: 75% (300) for train set and 25% (100) for test set
# X_train + X_test = X_true
# y_train + y_test = y_true
```

```
print(X_train)
```

```
print(y_train)
```

```
print(X_test)
```

```
print(y_test)
```

## Feature Scaling

```
# it is not totally necessary but it improve training
from sklearn.preprocessing import StandardScaler
sc = StandardScaler() # values between -3 and 3
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
print(X_train)
```

```
print(X_test)
```

## Training the Logistic Regression model on the Training set

```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

## Predicting a new result

```
new_data = [[30, 87000]] # comma separated values in an array
new_data = sc.transform(new_data) # values scaled like data sets
new_pred_prob = classifier.predict_proba(new_data) # probabilities
new_pred_bool = classifier.predict(new_data) # 0 (no) or 1 (yes)
```

```
print(new_pred_prob)
```

```
print(new_pred_bool)
```

## Predicting the Test set and comparing results

```
y_pred = classifier.predict(X_test)
y_prob = classifier.predict_proba(X_test)
np.set_printoptions(precision=3) # set floating point precision
y_pred_res = y_pred.reshape(len(y_pred), 1)
y_prob_res = y_prob.reshape(len(y_test), 2)
y_test_res = y_test.reshape(len(y_test), 1)
y_pred_vs_y_test = np.concatenate((y_pred_res, y_test_res, y_prob_res), 1)
print(y_pred_vs_y_test)
```

## Making the Confusion Matrix

```
from sklearn.metrics import confusion_matrix, accuracy_score
confusion = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
```

```
print('[[T0 F1') # 0 predictions: True 0 (pred 0), False 1 (pred 0)
print('  F0 T1]]') # 1 predictions: False 0 (pred 1), True 1 (pred 1)
print()
print(confusion)
# 0 (no) 65/73 correct predictions (8 wrong)
# 1 (yes) 24/27 correct predictions (3 wrong)
```

```
print(accuracy)
# 89% total of correct predictions (11% wrong)
```

## Visualising the Training set results

```
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 5, stop = X_set[:, 0].max() + 5, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 10000, stop = X_set[:, 1].max() + 10000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], color = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (train set)')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 5, stop = X_set[:, 0].max() + 5, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 10000, stop = X_set[:, 1].max() + 10000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], color = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (test set)')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
